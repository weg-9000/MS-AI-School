```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.set_printoptions(suppress=True,  # 지수 표기법 억제 (예: 1e-10 대신 0.0000000001로 표시)
                    precision=8)    # 소수점 이하 자리수를 8자리로 표현
```

# 데이터 준비

```python
# hitter1 = pd.read_csv('data/2000_2001_hitter.csv')
# hitter2 = pd.read_csv('data/2002_2013_hitter.csv')
# hitter3 = pd.read_csv('data/2014_hitter.csv')

hitter1 = pd.read_csv('https://raw.githubusercontent.com/JayoungKim-ai/ML_dataset/refs/heads/main/2000_2001_hitter.csv')
hitter2 = pd.read_csv('https://raw.githubusercontent.com/JayoungKim-ai/ML_dataset/refs/heads/main/2002_2013_hitter.csv')
hitter3 = pd.read_csv('https://raw.githubusercontent.com/JayoungKim-ai/ML_dataset/refs/heads/main/2014_hitter.csv')

print(f'hitter1 >>> {hitter1.shape}')
print(f'hitter2 >>> {hitter2.shape}')
print(f'hitter3 >>> {hitter3.shape}')

display(hitter1.head(3))
display(hitter2.head(3))
display(hitter3.head(3))
```

```python
# 데이터프레임 병합 
hitter_concat = pd.concat([hitter1, hitter2, hitter3], 
                          ignore_index=True) # 기존 인덱스 무시
hitter_concat.tail()
```

```python
# 변수 선택
X = hitter_concat[['OPS','ISO','SECA','TA','RC','RC/27','wOBA','XR']]
y = hitter_concat['YrPlayer']

print(X.shape, y.shape)
display(X.head())
display(y.head())
```

```python
# 데이터 분포 확인
X.plot(kind='box')
```

```python
# 데이터 스케일링
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X.loc[:,:] = scaler.fit_transform(X)
X.head()
```

```python
X.plot(kind='box');
```

# k-means clustering

### 최적의 k 찾기

```python
# 엘보우 기법으로 최적의 k 찾기
from sklearn.cluster import KMeans
inerita = []
for i in range(2,8):
    km = KMeans(n_clusters=i)
    km.fit(X)
    inerita.append(km.inertia_)

plt.plot(range(2,8), inerita, marker = "o");
```

### 군집화

```python
import silhouette_analysis as s
from sklearn.metrics import silhouette_score

km = KMeans(n_clusters=4, random_state=10)
km_cluster = km.fit_predict(X)

print(f'실루엣 점수: {silhouette_score(X, km_cluster)}')

s.silhouette_plot(X, 4)
```

### 군집화 결과 분석

```python
# 데이터프레임 생성
df = pd.DataFrame(X, columns= X.columns)
df['km_cluster'] = km_cluster
df.head()
```

```python
sns.pairplot(df,hue='km_cluster');
```

```python
# 군집의 데이터 특징 분석
cols = df.columns[:-1]

plt.figure(figsize=(15,8))
for i, col in enumerate(cols):
    plt.subplot(2,4,i+1)
    sns.boxplot(data = df , y= col, hue= 'km_cluster' , palette = 'muted')
    plt.title(col)


plt.tight_layout;

```

```python
# 개별 데이터 실루엣 계수 컬럼 추가
from sklearn.metrics import silhouette_samples
df['silhouette'] = silhouette_samples(X,km_cluster)
df.head()
```

```python
# 잘못 군집화 된 클러스터
df.loc[df['silhouette'] < 0]
```

```python
# 클러스터 별 중심 player
df['player'] = y
max_idx = df.groupby('km_cluster')['silhouette'].idxmax()
df.loc[max_idx, ['player','km_cluster']]
```

# 주성분분석

```python
# 모든 특성(컬럼)에 대해 주성분 계산하여 각 데이터포인트를 주성분 공간으로 변환
from sklearn.decomposition import PCA
p = PCA()
principal_components = p.fit_transform(X)
```

```python
# 컬럼별 설명된 분산산
p.explained_variance_ratio_
```

```python
# 컬럼별 설명된 분산 비율 확인 cumsum은 누적 분산비율을 확인 -> 분산 비율이 큰 데이터가 이 데이터를 잘 설명!!
p.explained_variance_ratio_.cumsum()
```

```python
# 2개의 주성분으로 차원 축소
from sklearn.decomposition import PCA
p = PCA(n_components=2)
principal_components = p.fit_transform(X)
principal_components
```

```python
# 주성분으로 군집화
k =4
km =KMeans(n_clusters=k, random_state=10)
pca_cluster=km.fit_predict(principal_components)

print(f'실루엣 점수:{silhouette_score(principal_components, pca_cluster)}')

s.silhouette_plot(principal_components, k)
```

```python
# 군집 결과 시각화
df_pca=pd.DataFrame(principal_components,columns=['pca1','pca2'])
df_pca['pca_cluster'] =pca_cluster
sns.scatterplot(data=df_pca, x='pca1', y='pca2', hue='pca_cluster', palette='muted');
```
