```python
import requests
import re
endpoint ="your endpoint"

api_key = "your api key"

ai_search_endpint = "your-search-endpoint"

ai_search_api_key = "your-search-api-key"

ai_search_index = "your-search-index"

ai_search_semantic = "your-search-semantic"

def chat(promt):

  headers = {
      "Content-Type": "application/json",
      "api-key": api_key 

  }

  body ={
    
          "messages": [
        {
          "role": "system",
          "content":"너는 남해 영행 전문가야"
        },
        {
          "role": "user",
          "content":promt
            }
          ],
          'temperature': 0.7,
          'top_p': 0.95,
          'max_tokens': 800,
          "data_sources": [
          {
            "type": "azure_search",
            "parameters": {
              "endpoint": ai_search_endpint,
              "index_name": ai_search_index,
              "semantic_configuration": ai_search_semantic,
              "query_type": "simple",
              "fields_mapping": {},
              "in_scope": True,
              "filter": None,
              "strictness": 3,
              "role_information": "남해 군청에서 일하는 직원이기 때문에 너는 남해에 대한 모든 정보를 알려줄수있어",
              "top_n_documents": 5,
              "authentication": {
                "type": "api_key",
                "key": ai_search_api_key
              },
              "key": ai_search_api_key,
            }
          }
          ]
  }

  response = requests.post(endpoint, headers=headers, json=body)

  if response.status_code == 200:
    response_json = response.json()
    messages = response_json['choices'][0]['message']
    citiation_list = messages['context']['citations']
    role = messages['role']
    content = messages['content']
    content = re.sub(r'\[doc(\d+)\]', r'[참조 \1]', content)
    print(role)
    return  content, citiation_list
  else:
    return ""

content, citiation_list = chat("남해 여행지 추천천")
from IPython.display import Markdown
display(Markdown(content))

for index in range(len(citiation_list)):
  c = citiation_list[index]
  print("Index : {}".format(index+1), c['content'])
```
